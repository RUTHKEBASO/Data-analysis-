Relationship between budget and box office, we will use scatter plot for visualization to show how one facter affects the other factor.


```python
df['Box_Office_Million_USD'].isnull().sum()
```

```python
df['Budget_Million_USD'].isnull().sum()
```

Due to a high percentage (~98%) of missing values in budget and box office data, the analysis is based on a limited subset of the dataset and may not fully represent overall trends.

```python
import matplotlib.pyplot as plt

# Keep only rows with both values
df_finance = df.dropna(subset=['Budget_Million_USD', 'Box_Office_Million_USD']) #created a variable df_finance and drop rows with missing values.
plt.figure(figsize=(10,10))

plt.scatter(df_finance['Budget_Million_USD'], df_finance['Box_Office_Million_USD'])

plt.title("Budget vs Box Office Revenue")
plt.xlabel("Budget")
plt.ylabel("Box Office")
plt.grid(True)

plt.show()
```

The scatter plot reveals several outliers where films with relatively low budgets generate high box office revenue, indicating strong return on investment. Conversely, there are a few high-budget films that do not achieve equally high revenue, suggesting underperformance. These patterns indicate that while there is a general positive relationship between budget and box office revenue, higher investment does not guarantee success, and other factors play a significant role in determining a film’s performance.

```python
import numpy as np

x = df_finance['Budget_Million_USD']
y = df_finance['Box_Office_Million_USD']

# line of best fit
m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(8,6))
plt.scatter(x, y)
plt.plot(x, m*x + b)

plt.title("Budget vs Box Office with Trend Line")
plt.xlabel("Budget")
plt.ylabel("Box Office")
plt.grid(True)

plt.show()
```

**World** **map**

```python

df_countries = df[['Country_Origin']].dropna()

# split multiple countries
df_countries['Country_Origin'] = df_countries['Country_Origin'].str.split('[,|]')

df_countries = df_countries.explode('Country_Origin')

# clean spaces
df_countries['Country_Origin'] = df_countries['Country_Origin'].str.strip()
```

Count production per country

```python
country_counts = df_countries['Country_Origin'].value_counts().reset_index()
country_counts.columns = ['country', 'count']
```

```python
import plotly.express as px

fig = px.choropleth(
    country_counts,
    locations="country",
    locationmode="country names",
    color="count",
    color_continuous_scale="Viridis",
    title="Global Animation Production by Country"
)

fig.show()
```

The world map reveals a strong concentration of animation film production in North America, Europe, and parts of Asia, particularly countries such as the United States, Japan, and China. In contrast, African countries show very limited animation production, indicating the lowest representation globally. This suggests a significant regional imbalance in the animation industry, where production is heavily dominated by a few regions.

***Animation History Timeline Visualization***

```python
import pandas as pd

# Placeholder for timeline data. Replace this with your actual data.
timeline_data = {
    'Year': [1878, 1908, 1928, 1937, 1960, 1995, 2001, 2010],
    'Event': [
        'First animated film (Praxinoscope)',
        'Fantasmagorie (Emile Cohl)',
        'Steamboat Willie (Disney)',
        'Snow White (First full-length)',
        'The Flintstones (First animated prime-time series)',
        'Toy Story (First full-length CGI)',
        'Shrek (Oscar for Best Animated Feature)',
        'Despicable Me (Minions introduced)'
    ]
}
timeline = pd.DataFrame(timeline_data)

plt.figure(figsize=(15,8)) # Increased figure size for better readability

# Scatter plot for the timeline points
plt.scatter(timeline['Year'], [1]*len(timeline), s=100, zorder=2) # zorder to ensure points are on top

# Annotate each point with its event. Adjust y-position and rotation for clarity.
for i, row in timeline.iterrows():
    plt.text(row['Year'], 1.02 + (i % 2 * 0.05), row['Event'], rotation=45, ha='left', fontsize=9, va='bottom') # Alternating y-pos

plt.yticks([]) # Remove y-axis ticks as they are not meaningful here
plt.xlabel("Year")
plt.title("Timeline of Animation History (1878 – Present)")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.ylim(0.95, 1.15) # Adjust y-limits to make space for annotations

plt.show()
```

The timeline visualization shows the evolution of animation from early motion studies in 1878 to modern AI-assisted animation. The industry shows slow initial progress, followed by rapid technological advancement from the 1990s onwards, particularly with the introduction of CGI and digital tools.
